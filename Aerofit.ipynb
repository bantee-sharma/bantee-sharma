3.Visual Analysis - Univariate & Bivariate

 (a)For continuous variable(s): Distplot, countplot, histogram for univariate analysis

```python
sns.histplot(df['Age'])
```

```python
sns.countplot(x=df["Gender"])
```

```python
#bivariate
#Categorical vs Numerical
sns.barplot(x = df['Product'], y = df['Income'])
```

```python
#Numerical vs Numerical
sns.scatterplot(x = df['Age'], y = df['Income'])
```

```python
#Categorical vs Categorical
sns.countplot(x = df['Product'], hue = df['Gender'])
```

(b) For categorical variable(s): Boxplot



```python
sns.boxplot(x= "Gender" ,y="Age",hue="Product",data=df)
```

```python
sns.heatmap(df.corr(), annot=True)
```
